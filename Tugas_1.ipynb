```python
print('Nim: 8020220216 -- 16')
from pylab import *
from PIL import Image
import numpy as np

image_sakura =Image.open('sakura-4141766_640.jpg')
print('\nGambar Asli')


image_rgb = image_sakura.convert('RGB')
rgb_value = np.array(image_rgb)

print('nilai pixel sebelum:')
print(rgb_value)

imshow(image_sakura)
show()

#Nilai R, G, dan B
r_value = rgb_value[:, :, 0]
g_value = rgb_value[:, :, 1]
b_value = rgb_value[:, :, 2]

#Mengubah nilai G dengan rumus G - 16
g_value16 = g_value - 16
g_value16 = np.clip(g_value16, 0, 255) #dalam rentang 0-255

rgb_value16 = np.stack((r_value, g_value16, b_value), axis = 2)
image_rgb16 = Image.fromarray(uint8(rgb_value16))

pixel_rgb16 = np.array(image_rgb16)

print('\nGambar yang sudah diubah nilai g')
print('Nilai pixel: ')
print(pixel_rgb16)
imshow(image_rgb16)
show()
```
