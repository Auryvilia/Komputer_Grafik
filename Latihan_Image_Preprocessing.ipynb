```python
#Properti Kontur

from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt

#membaca dan mempersiapkan gambar
drive.mount('/content/drive')
image_path = '/content/drive/MyDrive/Kuliah/Semester 5/Komputer Grafik/red-apple-on-table.jpg'
image = cv2.imread(image_path)
# Change this line to convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Menggunakan Canny edge Detection untuk menemukan tepi
edges = cv2.Canny(gray, 100, 200)

#Menemukan kontur dari hasil canny
contours, hierarchy = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

#Menghitung properti kontur
for contour in contours:
    # menghitung area (Luas dari area yang dibatasi oleh kontur)
    area = cv2.contourArea(contour)

    # Menghitung Perimeter (Keliling objek yang dibatasi kontur)
    perimeter = cv2.arcLength(contour, True)

    #Menghitung Centroid
    M = cv2.moments(contour) # Menghitung momen dari kontur
    if M["m00"] != 0: # M["m00"] adalah momen nol, yang digunakan untuk menghitung centroid (Titik pusat objek)
        cx = int(M["m10"] / M["m00"]) #koordinat X centroid
        cy = int(M["m01"] / M["m00"]) #koordinat Y centroid
    else:
        cx, cy = 0, 0

    # Menampilkan properti kontur / informasi kontur
    print(f"Area: {area}, Perimeter: {perimeter}, Centroid: ({cx}, {cy})")

    #Gambar centroid pada gambar
    cv2.circle(image, (cx, cy), 7, (255, 0, 0), -1) # Gambar titik centroid (merah)

#Menampilkan hasil
plt.imshow(cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
```

```python
#Pengenalan Bentuk dengan Kontur
#Menggunakan Threshold

from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt

#membaca dan mempersiapkan gambar
drive.mount('/content/drive')
image_path = '/content/drive/MyDrive/Kuliah/Semester 5/Komputer Grafik/geometri.jpg'
image = cv2.imread(image_path)
# Change this line to convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

#Menemukan kontur dari hasil threshold
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

#Menemukan bentuk berdasarkan jumlah sisi pada kontur
for contour in contours:
    # Aproksimasi kontur untuk menemukan poligon
    epsilon = 0.04 * cv2.arcLength(contour, True)
    approx = cv2.approxPolyDP(contour, epsilon, True)

    #Menentukan jenis bentuk
    if len(approx) == 3:
        shape_name = "Segitiga"
    elif len(approx) == 4:
      # Menentukan apakah persegi atau persegi panjang
        (x, y, w, h) = cv2.boundingRect(contour)
        aspect_ratio = float(w) / h
        if 0.95 < aspect_ratio < 1.05:
            shape_name = "Bujur Sangkar"
        else:
            shape_name = "Persegi Panjang"
    else:
        shape_name = "Lingkaran"

    # Gambar bentuk dan nama pada gambar
    cv2.drawContours(image,[approx], -1, (0, 255, 0), 2)
    cv2.putText(image, shape_name, (approx [0] [0] [0], approx [0] [0] [1] - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255),2)


#Menampilkan Hasil
plt.imshow(cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
plt.axis('off')
plt.show()
```
