**Image Contours and Histograms**

```python
from PIL import Image
from pylab import *
# read image to array
img_cantik = array(Image.open("studio-ghibli-7415572_640.jpg").convert("L"))
# create a new figure
figure()
# don’t use colors
gray()
# show contours with origin upper left corner
contour(img_cantik, origin="image")
axis("equal")
axis("off")
figure()
hist(img_cantik.flatten(),128)
show()
```
